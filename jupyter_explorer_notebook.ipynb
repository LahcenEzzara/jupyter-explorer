```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
# Histogram of sepal length
plt.figure(figsize=(8, 6))
plt.hist(data['sepal length (cm)'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of sepal length vs sepal width
plt.figure(figsize=(8, 6))
plt.scatter(data['sepal length (cm)'], data['sepal width (cm)'], color='green')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Box plot of petal length
plt.figure(figsize=(8, 6))
data.boxplot(column='petal length (cm)')
plt.title('Box Plot of Petal Length')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Line plot of mean sepal length by species
plt.figure(figsize=(8, 6))
data.groupby('species')['sepal length (cm)'].mean().plot(kind='line', marker='o', color='red')
plt.title('Mean Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length (cm)')
plt.grid()
plt.show()
```
